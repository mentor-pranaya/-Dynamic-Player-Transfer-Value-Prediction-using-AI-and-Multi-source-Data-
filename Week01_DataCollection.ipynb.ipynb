```python
import pandas as pd
import json
import os

# Base folder (adjust if your path is different)
BASE_PATH = r"C:\Users\M.ANTONY ROJES\Downloads\Infosys\open-data-master\data"

# Load competitions.json
def load_competitions():
    path = os.path.join(BASE_PATH, "competitions.json")
    return pd.read_json(path)

# Load matches.json for a given competition
def load_matches(country, competition, season):
    path = os.path.join(BASE_PATH, "matches", country, competition, season, "matches.json")
    return pd.read_json(path)

# Load events.json for a given match_id
def load_events(match_id):
    path = os.path.join(BASE_PATH, "events", f"{match_id}.json")
    with open(path, "r", encoding="utf-8") as f:
        return pd.json_normalize(json.load(f))

# Load lineups.json for a given match_id
def load_lineups(match_id):
    path = os.path.join(BASE_PATH, "lineups", f"{match_id}.json")
    with open(path, "r", encoding="utf-8") as f:
        return pd.json_normalize(json.load(f))
```

```python
import pandas as pd
import os

# <-- change this to where you saved it
INJURY_PATH = r"C:\Users\M.ANTONY ROJES\Downloads\Infosys\open-data-master\player_injuries_impact.csv"

assert os.path.exists(INJURY_PATH), "Injury CSV path is wrong. Fix INJURY_PATH."
```

```python
inj = pd.read_csv(
    INJURY_PATH,
    encoding="utf-8-sig",
    parse_dates=["Date of Injury", "Date of return"],
    dayfirst=False,  # your dates look like 'Nov 9, 2019' (US style)
    infer_datetime_format=True
)

inj.head()
```

```python
print("Shape:", inj.shape)
print("\nColumns:", inj.columns.tolist())

print("\nNulls (top 15):")
print(inj.isna().sum().sort_values(ascending=False).head(15))

print("\nUnique players:", inj["Name"].nunique())
print("Unique teams:", inj["Team Name"].nunique())
print("Seasons:", inj["Season"].unique())

print("\nTop 10 injury types:")
print(inj["Injury"].value_counts().head(10))
```

```python
inj["Date of Injury"] = pd.to_datetime(inj["Date of Injury"], errors="coerce")
inj["Date of return"] = pd.to_datetime(inj["Date of return"], errors="coerce")

```

```python
inj["days_out"] = (inj["Date of return"] - inj["Date of Injury"]).dt.days
inj["days_out"] = inj["days_out"].clip(lower=0)  # remove negatives
```

```python
print(inj["days_out"].describe())
```

```python
injury_by_player = (
    inj.groupby("Name", as_index=False)
       .agg(
           injuries=("Injury","count"),
           total_days_out=("days_out","sum"),
           teams=("Team Name", lambda x: ", ".join(sorted(set(map(str, x)))))
       )
       .sort_values(["injuries","total_days_out"], ascending=[False, False])
)
injury_by_player.head(10)

```

```python
injury_by_season = (
    inj.groupby(["Season","Name"], as_index=False)
       .agg(
           injuries=("Injury","count"),
           total_days_out=("days_out","sum")
       )
       .sort_values(["Season","injuries"], ascending=[True, False])
)
injury_by_season.head(10)
```

```python
inj.to_csv("injuries_raw.csv", index=False)
injury_by_player.to_csv("injuries_by_player.csv", index=False)
injury_by_season.to_csv("injuries_by_season.csv", index=False)

print("Saved: injuries_raw.csv, injuries_by_player.csv, injuries_by_season.csv ✅")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.transfermarkt.co.in/spieler-statistik/marktwertaenderungen/marktwertetop?spieler_land_id=0&verein_land_id=0&position=alle&wettbewerb_id=&minMarktwert=100000&maxMarktwert=200000000&minAlter=16&maxAlter=45"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

players = []
table = soup.find("table", class_="items")

for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) >= 5:
        name = cols[1].get_text(strip=True)
        age = cols[2].get_text(strip=True)
        nationality = cols[3].img['title'] if cols[3].img else ""
        club = cols[4].img['alt'] if cols[4].img else ""
        market_value = cols[5].get_text(strip=True)
        players.append([name, age, nationality, club, market_value])

df = pd.DataFrame(players, columns=["Name", "Age", "Nationality", "Club", "Market Value"])
df.to_csv("transfermarkt_market_values.csv", index=False)
print("CSV file saved successfully.")
```

```python
import pandas as pd

df = pd.read_csv("transfermarkt_market_values.csv")
print(df.head())  # Shows the first 5 rows
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure all values are treated as strings before cleaning
df['Market Value'] = df['Market Value'].astype(str)

# Remove currency symbols and units
df['Market Value'] = df['Market Value'].str.replace('€', '', regex=False)
df['Market Value'] = df['Market Value'].str.replace('m', '', regex=False)
df['Market Value'] = df['Market Value'].str.replace('-', '', regex=False)  # Handle missing values

# Convert to float, coercing errors to NaN
df['Market Value'] = pd.to_numeric(df['Market Value'], errors='coerce')

# Drop rows with NaN market values
df = df.dropna(subset=['Market Value'])

# Plot the histogram
plt.hist(df['Market Value'], bins=20, color='royalblue', edgecolor='white')
plt.title("Player Market Value Distribution")
plt.xlabel("Market Value (€ millions)")
plt.ylabel("Number of Players")
plt.grid(True)
plt.tight_layout()
plt.show()
```
