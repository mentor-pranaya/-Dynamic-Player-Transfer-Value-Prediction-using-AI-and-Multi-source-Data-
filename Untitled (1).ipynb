```python
# 📌 Week 1 - Data Collection & Initial Exploration (Matplotlib Only)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Dataset
file_path = "C:/Users/Khwaish Goel/OneDrive/Desktop/players_22.csv"
df = pd.read_csv(file_path)

print("✅ Dataset loaded successfully!")
print("Shape of dataset:", df.shape)

# 2. Preview Dataset
print("\nFirst 5 rows of dataset:\n", df.head())

print("\nDataset Info:\n")
print(df.info())

# 3. Missing Values & Duplicates
print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))

print("\nDuplicate rows:", df.duplicated().sum())

# 4. Summary Statistics
print("\nSummary Statistics (Numerical Columns):")
print(df.describe())

# 5. Distributions of Key Columns
key_cols = ["age", "height_cm", "weight_kg", "overall", "potential", "value_eur", "wage_eur"]
for col in key_cols:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        plt.hist(df[col].dropna(), bins=30, edgecolor='black')
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()

# 6. Categorical Exploration
if "nationality_name" in df.columns:
    print("\nTop 10 Nationalities:\n", df['nationality_name'].value_counts().head(10))

if "club_name" in df.columns:
    print("\nTop 10 Clubs:\n", df['club_name'].value_counts().head(10))

# 7. Outlier Analysis (Top Players)
if "value_eur" in df.columns:
    print("\nTop 5 Most Valuable Players:")
    print(df[['short_name','age','overall','potential','value_eur']]
          .sort_values('value_eur', ascending=False).head(5))

if "wage_eur" in df.columns:
    print("\nTop 5 Highest Wage Players:")
    print(df[['short_name','age','overall','potential','wage_eur']]
          .sort_values('wage_eur', ascending=False).head(5))

# 8. Correlation Heatmap (using matplotlib only)
corr = df[key_cols].corr()
fig, ax = plt.subplots(figsize=(10,6))
cax = ax.matshow(corr, cmap='coolwarm')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar(cax)
plt.title("Correlation Heatmap of Key Features", pad=20)
plt.show()

# 9. Scatterplots for Insights
if "value_eur" in df.columns and "overall" in df.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(df["overall"], df["value_eur"], alpha=0.5)
    plt.title("Overall Rating vs Market Value")
    plt.xlabel("Overall")
    plt.ylabel("Market Value (EUR)")
    plt.show()

if "value_eur" in df.columns and "age" in df.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(df["age"], df["value_eur"], alpha=0.5)
    plt.title("Age vs Market Value")
    plt.xlabel("Age")
    plt.ylabel("Market Value (EUR)")
    plt.show()

```

```python
import tweepy
import pandas as pd
import time
import os

# --- Authenticate with Twitter ---
bearer_token = "AAAAAAAAAAAAAAAAAAAAAIrg3gEAAAAAezJatO1JKBV9FMGL0K1%2FgS7F21o%3DWIpCHaiWktaGoPPplpvcHPBRFTRWhm2gJJcWJeDEUSTaf2S07U"  # paste your valid bearer token
client = tweepy.Client(bearer_token=bearer_token, wait_on_rate_limit=True)

# --- Load FIFA dataset ---
file_path = "C:/Users/Khwaish Goel/OneDrive/Desktop/players_22.csv"
df = pd.read_csv(file_path, low_memory=False)

# --- Select top 10 players by market value ---
top_players = df.sort_values("value_eur", ascending=False).head(10)

# --- Output CSV file ---
output_file = "top10_players_tweets.csv"

# --- Fetch tweets for each player ---
for name in top_players["short_name"].tolist():
    # clean name → use last word (e.g. "Messi", "Ronaldo")
    clean_name = name.split(" ")[-1]
    query = f'"{clean_name}" football -is:retweet lang:en'
    
    try:
        tweets = client.search_recent_tweets(
            query=query, 
            max_results=20,
            tweet_fields=["id", "text", "created_at", "lang"]
        )
        
        if tweets.data:
            df_player = pd.DataFrame([{
                "player": name,
                "query_used": query,
                "tweet_id": t.id,
                "text": t.text,
                "created_at": t.created_at,
                "lang": t.lang
            } for t in tweets.data])

            # append results to CSV file
            df_player.to_csv(output_file, mode="a", index=False,
                             header=not os.path.exists(output_file), encoding="utf-8")
            print(f"✅ Saved {len(df_player)} tweets for {name}")
        else:
            print(f"⚠️ No tweets found for {name} (query: {query})")
    
    except Exception as e:
        print(f"❌ Error for {name}:", e)
    
    time.sleep(2)  # pause to reduce rate limit risk

print("📂 Finished! All tweets saved in:", output_file)

# --- Preview saved file ---
tweets_df = pd.read_csv(output_file)
print("\n🔍 Preview of collected tweets:\n", tweets_df.head())


```
